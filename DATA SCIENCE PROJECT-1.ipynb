### SALARY PARSING

```python
df=df[df['Salary Estimate']!='-1']#Removing -1 from salary estimate
```

```python
df.shape
```

```python
df.head(20)
```

Now removing (Glassdoor est.) from Salary Estimate

```python
df['hourly']=df['Salary Estimate'].apply(lambda x:1 if 'per hour' in x.lower() else 0)
```

```python
df['employer_provided']=df['Salary Estimate'].apply(lambda x:1 if 'employer provided salary' in x.lower() else 0) 
```

```python
salary=df['Salary Estimate'].apply(lambda x:x.split('(')[0])
```

```python
minus_kd=salary.apply(lambda x:x.replace('K','').replace('$',''))
```

```python
minus_kd
```

Now we have removed the GlassDoor Est. from Salary Estimate..Now There are some Salaries with 'Per Hour' or 'Employer Paid' Tag..We have to seperate them both

```python
min_hr=minus_kd.apply(lambda x:x.lower().replace('per hour','').replace('employer provided salary:',''))
```

```python
df['min_salary']=min_hr.apply(lambda x:int(x.split('-')[0]))
```

```python
df['max_salary']=min_hr.apply(lambda x:int(x.split('-')[1]))
```

```python
df['employer_provided']=df['Salary Estimate'].apply(lambda x:1 if 'Employer Provided Salary:' in x.lower() else 0)
```

```python
df.head()
```

```python
df['avg_salary']=(df['min_salary']+df['max_salary'])/2
```

```python
df.head()
```

### COMPANY NAME TEXT ONLY

```python
df['company text']=df['Company Name'].apply(lambda x:x[:-5])
```

```python
df.head()
```

### STATE FIELD

```python
df['job_state']=df['Location'].apply(lambda x:x.split(',')[1])
```

```python
df.head()
```

```python
df['job_state'].value_counts()#number of jobs in every state
```

```python
df['same_state']=df.apply(lambda x:1 if x.Location==x.Headquarters else 0,axis=1)
```

```python
df.head()
```

### AGE OF COMPANY

```python
df['age']=df['Founded'].apply(lambda x:x if x<1 else 2020-x)
```

```python
df.head()
```

## EDA(Exploratory Data Analysis)

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'mle'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'
```

```python
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr.' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
        return 'senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    else:
        return 'na'
```

```python
salary['job_simplified']=salary['Job Title'].apply(title_simplifier)
```

```python
salary['job_simplified'].value_counts()
```

```python
salary['seniority']=salary['Job Title'].apply(seniority)
```

```python
salary['seniority'].value_counts()
```

```python
salary['job_state'].value_counts()
```

We have to fix the 'los angeles' and remove the space in between

```python
salary['job_state']=salary['job_state'].apply(lambda x:x.strip() if x.strip().lower()!='los angeles'
                                     else 'CA')
```

```python
salary['job_state'].value_counts()
```

```python
#Job Desription Length
salary['desc_len']=salary['Job Description'].apply(lambda x:len(x))
```
